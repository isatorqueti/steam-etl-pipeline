```python
import json
import pandas as pd
from pathlib import Path
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
path_ranking = '../data/steam_ranking.json'

with open(path_ranking) as f:
    data_ranking = json.load(f)

df_ranking = pd.json_normalize(data_ranking['response']['ranks'])
df_ranking.head()
```

```python
path_apps = '../data/steam_apps.json'

with open(path_apps) as f:
    data_apps = json.load(f)

df_apps = pd.json_normalize(data_apps['applist']['apps'])
df_apps.head()
```

```python
df_final = df_ranking.merge(df_apps, on='appid', how='left')
df_final[['appid', 'name', 'concurrent_in_game']].head(10)
```

```python
print("Tipo no Ranking:")
print(df_ranking['appid'].dtype)

print("Tipo nos Apps:")
print(df_apps['appid'].dtype)
```
